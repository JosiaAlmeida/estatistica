```python
import numpy as np
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

```python
dados = norm.rvs(size = 1000)
dados
```

```python
plt.hist(dados, bins = 20)
plt.title("Dados")
```

```python
#Gerar grafico para verificar se a distribuicao  normal

fig, ax = plt.subplots()
stats.probplot(dados, fit=True, plot=ax)
plt.show()
```

```python
#Teste de shapiro
#Segundo argumento  o valor de p, permitindo hipoteses nula
stats.shapiro(dados)
```

```python
dados2 = skewnorm.rvs(4, size=1000)
dados2
```

```python
plt.hist(dados2, bins = 20)
plt.title("Dados")
```

```python

fig, ax = plt.subplots()
stats.probplot(dados2, fit=True, plot=ax)
plt.show()
```

```python
#Teste de shapiro
#Segundo argumento  o valor de p, recusando hipoteses nula
stats.shapiro(dados2)
```
